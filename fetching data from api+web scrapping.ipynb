```python
containers=soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'})
```

```python
name=[]
price=[]
specs=[]
sim=[]
battery=[]
networks=[]
screen_size=[]
processor=[]
camera=[]
ram=[]
os=[]
for i in containers:
 
    name.append(i.find_all('h2')[0].text)
   
    price.append(i.find_all('span',{'class':'price'})[0].text.replace(',',''))
    
       
    x=i.find_all('div',{'class':'score rank-2-bg'})
    for w in x:
        if w:
            specs.append(w.find('b').text)
    y=i.find_all('div',{'class':'score rank-1-bg'})
    for j in y:

        specs.append(i.find('b').text)
    z=i.find_all('div',{'class':'score rank-3-bg'})
    for k in z:

        specs.append(i.find('b').text)
            

    
    m=i.find_all('ul',{'class':'sm-feat specs'})
    for u in m:

           
        sim.append(u.find_all('li')[0].text)
            
   
  
        networks.append(u.find_all('li')[1].text)
       
        processor.append(u.find_all('li')[2].text)

        ram.append(u.find_all('li')[3].text)
        
        battery.append(u.find_all('li')[4].text)
        
        screen_size.append(u.find_all('li')[5].text)
       
        camera.append(u.find_all('li')[6].text)
       
        os.append(u.find_all('li')[7].text)
        
specs.insert(198,77) 
specs.insert(199,80)
df1=pd.DataFrame({'name':name,
                  'price':price,
                'specs':specs,
                'sim':sim,
                'battery':battery,
                'ram':ram,
                'networks':networks,
                'processor':processor,
                 'screen_size':screen_size,
                 'os':os,
                 'camera':camera
                })
```
